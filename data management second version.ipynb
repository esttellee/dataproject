# Scraping des infos du top 250

```python
titles = soup.find_all('td', class_='titleColumn')
```

essayer de ne faire apparaître que les acteurs présents dans la liste de allociné

```python
#récupere la liste des acteurs connus
list_actors_famous = []
for page in range(1, 21):
    url_actor_star = f'https://www.allocine.fr/personne/top/les-plus-vues/ever/?page={page}'
#le f sert à formater l'URL
    response_actor = requests.get(url_actor_star)
    actor_soup = BeautifulSoup(response_actor.content, 'html.parser')
    actors_famous = actor_soup.find_all('a', class_= "meta-title-link")
    
#mettre tous les acteurs scrapper dans une liste   
    for actor in actors_famous:
        actor_star = actor.text.strip()
        list_actors_famous.append(actor_star)
```

```python
for title in titles:
    movie_title = title.a.text
    movie_year = title.span.text.strip('()')
    ratings = title.find_next_sibling('td', class_= 'ratingColumn')
    
 #aller scrapper sur les pages de chaque film
    movie_url = 'https://www.imdb.com' + title.a['href']
    req = Request(
    url=movie_url, 
    headers={'User-Agent': 'Mozilla/5.0'}
    )
    
    movie_response = urlopen(req).read()
    movie_soup = BeautifulSoup(movie_response, 'html.parser')
    genre_span = movie_soup.find('span', {'class': 'ipc-chip__text'}).text.strip()
        
    director_movie = movie_soup.find('a', class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").text.strip()
    
    actors_list = movie_soup.find_all("a", class_="sc-bfec09a1-1 fUguci")
    #créer une liste pour les acteurs en commun avec allociné
    actors_common = []   
    #essayer de n'afficher que les acteurs célèbres
    for actor in actors_list:
        actor_name = actor.text.strip()
        if actor_name in list_actors_famous:
            actors_common.append(actor_name)

    
    movie_dict = {
        'title': movie_title,
        'year': movie_year,
        'rating': float(ratings.text),
        'genre':genre_span,
        'director': director_movie,
        'actor': actors_common
    }
 
    print(movie_dict)
```

# visualisation

```python
import matplotlib.pyplot as plt

# Création d'une liste de titres de films et de leurs notes
titles = [movie['title'] for movie in movie_dict]
ratings = [movie['rating'] for movie in movie_dict]

# Création d'un graphique en barres horizontales pour les notes des films
plt.barh(titles, ratings)
plt.xlabel('Note')
plt.ylabel('Titre du film')
plt.title('Notes des films')
plt.show()

# Création d'un graphique en camembert pour les genres de films
genres = {}
for movie in movie_dict:
    genre = movie['genre']
    if genre in genres:
        genres[genre] += 1
    else:
        genres[genre] = 1

genre_labels = list(genres.keys())
genre_counts = list(genres.values())

plt.pie(genre_counts, labels=genre_labels, autopct='%1.1f%%')
plt.title('Genres de films')
plt.show()

# Création d'un graphique à barres pour les réalisateurs de films
directors = {}
for movie in movie_dict:
    director = movie['director']
    if director in directors:
        directors[director] += 1
    else:
        directors[director] = 1

director_labels = list(directors.keys())
director_counts = list(directors.values())

plt.bar(director_labels, director_counts)
plt.xlabel('Réalisateur')
plt.ylabel('Nombre de films')
plt.title('Réalisateurs de films')
plt.xticks(rotation=90)
plt.show()

# Création d'un graphique en nuage de points pour les acteurs communs avec Allociné
actors = {}
for movie in movie_dict:
    for actor in movie['actor']:
        if actor in actors:
            actors[actor] += 1
        else:
            actors[actor] = 1

actor_labels = list(actors.keys())
actor_counts = list(actors.values())

plt.scatter(actor_labels, actor_counts)
plt.xlabel('Acteur')
plt.ylabel('Nombre de films')
plt.title('Acteurs communs avec Allociné')
plt.xticks(rotation=90)
plt.show()
```
